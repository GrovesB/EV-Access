```python
import pandas as pd
from sqlalchemy import create_engine, inspect
import numpy as np
```

```python
ev_ranges_file = "Data/10963_EV_range_efficiency_1-6-22.xlsx"
ev_ranges_df = pd.read_excel(ev_ranges_file, skiprows=2)
```

```python
ev_ranges_df2 = ev_ranges_df.drop(ev_ranges_df.columns[[0,5,6,7,8,9]], axis=1)
```

```python
ev_ranges_df3 = ev_ranges_df2.dropna()
```

```python
ev_ranges_df3
```

```python
ev_gbym = ev_ranges_df3.groupby("Model Year").mean().reset_index()
```

```python
ev_gbymax = ev_ranges_df3.groupby("Model Year")["Range"].max().reset_index()
```

```python
ev_gbymax
```

```python
ev_gbym
```

```python
ev_ranges_by_year = pd.merge(ev_gbym,ev_gbymax, how="left", on="Model Year")
```

```python
ev_ranges_by_year
```

```python
ev_ranges_by_year.rename(columns={'Range_x' : 'Average Range', 'Range_y' : 'Max Range'}, inplace=True)
```
