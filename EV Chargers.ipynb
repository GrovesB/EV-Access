```python
import pandas as pd
from flask import Flask, render_template, jsonify
from flask_pymongo import PyMongo
import json
import ev_chargers
```

```python
def chargers():

    ev_chargers_file = "Data/EV_Charge.csv"
    ev_chargers_df = pd.read_csv(ev_chargers_file)
    
    #drop NAN fields
    ev_chargers_df.dropna(inplace=True)
    
    #group fields
    ev_location = ev_chargers_df.groupby(["Latitude", "Longitude"]).first()
    
    ev_location.reset_index
    
    return ev_location.to_dict('split')
```

```python
app = Flask(__name__)
```

```python
conn = "mongodb://localhost:27017/ev_location"
mongo = PyMongo(app, uri = conn)
```

```python
ev_locations = chargers()
```
